# 구현(Implementation)
구형(Implementation)이란 '머릿속에 있는 알고리즘을  소스코드로 바꾸는 과정'이다. 어떤 문제를 풀든 간에 소스코드를 작성하는 과정은 필수이기에 구현 문제 유형은 모든 범위의 코딩 테스트 문제 유형을 포함하는 개념이다.
알고리즘 문제를 해결할 때, 문제를 읽고 문제 풀이 방법을 고민한다. 고민 끝에 문제에 대한 정확한 풀이 방법이 떠오르면 정답 처리를 받을 수 있을까? 그렇지 않다. 생각해낸 문제 풀이 방법을 우리가 원하는 프로그래밍 언어로 정확히 구현해냈을 때 비로소 정답 처리를 받을 수 있다.
흔히 문제 해결 분야에서 구현 유형의 문제는 '풀이를 떠올리는 것은 쉽지만 소스코드로 옮기기 어려운 문제'를 의미한다. 알고리즘은 간단한데 코드가 지나칠 만큼 길어지는 문제, 특정 소수점 자리까지 출력해야 하는 문제, 문자열이 입력으로 주어졌을 때 한 문자 단위로 끊어서 리스트에 넣어야 하는 문제 등이 까다로운 구현 유형의 문제라고 할 수 있다. 대체로 사소한 조건 설정이 많은 문제일수록 코드로 구현하기 까다롭다.
완전 탐색, 시뮬레이션 유형을 모두 구현 유형으로 묶어서 다룰 수 있다.
    - 완전 탐색은 모든 경우의 수를 주저 없이 다 계산하는 해결 방법을 의미한다.
    - 시뮬레이션은 문제에서 제시한 알고리즘을 한 단계씩 차례대로 직접 수행해야 하는 문제 유형을 의미한다.

```python
# ex 4-1 up down left right

"""
여행가 A는 N X N 크기의 정사각형 공간위에 서 있다. 이 공간은 1 X 1 크기의 정사각형으로 나누어져 있다. 가장 왼쪽 위 좌표는 (1, 1)이며,
가장 오른쪽 아래 (N, N)에 해당한다. 여행가 A는 상, 하, 좌, 우 방향으로 이동할 수 있으며, 시작 좌표는 항상 (1, 1)이다.
계획서에는 하나의 줄에 띄어쓰기를 기준으로 하여 L, R, U, D 중 하나의 문자가 반복적으로 적혀 있다.
이때 여행가 A가 N X N 크기의 정사각형 공간을 벗어나는 움직임은 무시된다.
계획서가 주어졌을 때 여행가 A가 최종적으로 도착할 지점의 좌표를 출력하는 프로그램을 작성하라

입력 조건
    - 첫째 줄에 공간의 크기를 나타내는 N이 주어진다.
    - 둘째 줄에 여행가 A가 이동할 계획서 내용이 주어진다.

출력 조건
    - 첫째 줄에 여행가 A가 최종적으로 도착할 좌표를 공백으로 구분하여 출력한다.
"""

N = int(input(""))
move_quest = input("")

move_list = move_quest.split()

current_corr = [1, 1]

for i in move_list:
    if i == 'L':
        if current_corr[1] == 1:
            pass
        else:
            current_corr[1] = current_corr[1] - 1
    
    elif i == 'R':
        if current_corr[1] == N:
            pass
        else:
            current_corr[1] = current_corr[1] + 1
    
    elif i == 'U':
        if current_corr[0] == 1:
            pass
        else:
            current_corr[0] = current_corr[0] - 1
    
    elif i == 'D':
        if current_corr[0] == N:
            pass
        else:
            current_corr[0] = current_corr[0] + 1

print("{} {}" .format(current_corr[0], current_corr[1]))
```

```python
# ex 4-2 clock visualization

"""
정수 N이 입력되면 00시 00분 00초부터 N시 59분 59초까지 모든 시각 중에서 3이 하나라도 포함되는 모든 경우의 수를 구하는 프로그램
예를 들어 1을 입력했을 때 다음은 3이 하나라도 포함되어 있으므로 세어야 하는 시각이다.
    EX) 00시 00분 03초, 00시 13분 30초 ... etc

반면에 다음은 3이 하나도 포함되어 있지 않으므로 세면 안되는 시각이다.
    EX) 00시 02분 55초 ... etc
    
입력 조건
    - 첫째 줄에 정수 N이 입력된다.

출력 조건
    - 00시 00분 00초부터 N시 59분 59초까지 모든 시각 중에서 3이 하나라도 포함되는 모든 경우의 수
"""

N = int(input(""))

possible_list = []

for i in range(0, N+1):
    time_str = ''
    
    if i >= 10:
        hour_str = str(i)
    else:
        hour_str = '0' + str(i)
    
    for j in range(0, 60):
        if j >= 10:
            min_str = str(j)
        else:
            min_str = '0' + str(j)
        
        for k in range(0, 60):
            if k >= 10:
                sec_str = str(k)
            else:
                sec_str = '0' + str(k)
            
            time_str = hour_str + min_str + sec_str
            
            possible_list.append(time_str)

count_case = 0

for i in possible_list:
    if '3' in i:
        count_case += 1

print(count_case)
```

```python
# exercise proble. 왕실의 나이트

"""
정원은 8 x 8 좌표 평면이다. 정원의 특정한 한 칸에 나이트가 서 있다.
나이트는 말을 타고 있기에 이동할 때는 L자 형태로만 이동할 수 있으며 정원 밖으로는 나갈 수 없다.
나이트는 특정한 위치에서 다음과 같은 두가지 경우로 이동할 수 있다.
    - 수평으로 두칸 이동 후 수직으로 한 칸 이동
    - 수직으로 두칸 이동 후 수평으로 한 칸 이동

8 X 8 좌표 평면상에서 나이트의 위치가 주어졌을 때 나이트가 이동할 수 있는 경우의 수를 출력하는 프로그램을 작성하라.

입력 조건
    - 첫째 줄에 8 x 8 좌표 평면상에서 현재 나이트가 위치한 곳의 좌표를 나타내는 두 문자로 구성된 문자열이 입력된다.
        입력 문자는 a1처럼 열 + 행으로 이루어지낟.

출력 조건
    - 첫째 줄에 나이트가 이동할 수 있는 경우의 수를 출력하라
"""

knight_yx = input("")

knight_y_str = knight_yx[0]

if knight_y_str == 'a':
    knight_y = 1
elif knight_y_str == 'b':
    knight_y = 2
elif knight_y_str == 'c':
    knight_y = 3
elif knight_y_str == 'd':
    knight_y = 4
elif knight_y_str == 'e':
    knight_y = 5
elif knight_y_str == 'f':
    knight_y = 6
elif knight_y_str == 'g':
    knight_y = 7
elif knight_y_str == 'h':
    knight_y = 8
    
knight_x = int(knight_yx[1])

# x direction +2 -2 +1 -1, y direction +2 -2 +1 -1
# total 8 cases
consideration = [2, -2, 1, -1]

possible_count = 0

for del_x in consideration:
    result_x = knight_x    
    result_x += del_x
    
    if abs(del_x) == 2:
        del_y_list = [-1, 1]
        
        for del_y in del_y_list:
            result_y = knight_y
            result_y += del_y

            if ((result_x >= 1) & (result_x <= 8)) & ((result_y >= 1) & (result_y <= 8)):
                possible_count += 1
            else:
                pass
    
    elif abs(del_x) == 1:
        del_y_list = [-2, 2]
        
        for del_y in del_y_list:
            result_y = knight_y
            result_y += del_y

            if ((result_x >= 1) & (result_x <= 8)) & ((result_y >= 1) & (result_y <= 8)):
                possible_count += 1
            else:
                pass
    

print(possible_count)
```

```python
# exercise problem. 게임 개발

"""
게임 캐릭터가 맵 안에서 움직이는 시스템을 개발중이다.
캐릭터가 있는 장소는 1x1 크기의 정사각형으로 이루어진 nxm 크기의 직사각형으로, 각각의 칸은 육지 혹은 바다이다.
캐릭터는 동서남북 중 한 곳을 바라본다.
맵의 각 칸은 (A, B)로 나타낼 수 있고, A는 북쪽으로부터 떨어진 칸의 갯수, B는 서쪽으로 부터 떨어진 칸의 갯수이다.
캐릭터는 상하좌우로 움직일 수 있고, 바다로 되어 있는 공간에는 갈 수 없다.

    - 현재 위치에서 현재 방향을 기준으로 왼쪽 방향(반시계 방향으로 90도 회전한 방향)부터 차례대로 갈 곳을 정한다.
    - 캐릭터의 바로 왼쪽 방향에 아직 가보지 않은 칸이 존재한다면, 왼쪽 방향으로 회전한 다음 왼쪽으로 한 칸을 전진한다.
        왼쪽 방향에 가보지 않은 칸이 없다면, 왼쪽 방향으로 회전만 수행하고 1단계로 돌아간다.
    - 만약 네 방향 모두 이미 가본 칸이거나 바다로 되어 있는 칸인 경우, 바라보는 방향을 유지한채로 한 칸 뒤로 가고 1단계로 돌아간다.
        이 때 뒤쪽 칸이 바다라 뒤로 갈 수 없는 경우에는 움직임을 멈춘다.
위의 메뉴얼대로 캐릭터를 이동시킨 후, 캐릭터가 방문한 칸의 수를 출려갛는 프로그램을 만들어라

입력 조건
    - 첫째 줄에 맵의 크기를 n, m으로 공백으로 구분하여 입력한다.
    - 둘째 줄에 게임 캐릭터가 있는 좌표와 바라보는 방향을 서로 공백으로 구분하여 입력한다.
    - 셋째 줄부터 맵이 육지인지 바다인지에 대한 정보가 주어진다. (0 : 육지, 1 : 바다)
    - 처음 게임 캐릭터가 위치한 칸의 상태는 항상 육지이다.

출력 조건
    - 첫째 줄에 이동을 마친 후 캐릭터가 방문한 칸의 수를 출력한다.
"""
import numpy as np

first_str = input("")
second_str = input("")

N = int(first_str.split()[0])
M = int(first_str.split()[1])

current_x = int(second_str.split()[0])
current_y = int(second_str.split()[1])
current_dir = int(second_str.split()[2])

array_xy = np.zeros((N, M))

for i in range(0, N):
    input_str = input("")
    input_list = list(map(int, input_str.split()))
    
    array_xy[i] = input_list

current_position = [current_y, current_x]
move_history = array_xy.copy()

move_history[current_y][current_x] = 2

def rotate_left(current_dir):
    forhead_dir = (current_dir - 1) % 4
    
    return forhead_dir

def search_current_dir_move_ability(current_dir, current_position, array_xy, move_history):
    if current_dir == 0:
        # to the north
        forhead_x = current_position[1]
        forhead_y = current_position[0] - 1
    elif current_dir == 1:
        # to the east
        forhead_x = current_position[1] + 1
        forhead_y = current_position[0]
    elif current_dir == 2:
        # to the south
        forhead_x = current_position[1]
        forhead_y = current_position[0] + 1
    elif current_dir == 3:
        # to the west
        forhead_x = current_position[1] - 1
        forhead_y = current_position[0]
    
    move_ability = False
    
    if move_history[forhead_y][forhead_x] == 0:
        move_ability = True
    
    return move_ability

def move_to_currnet_dir(current_dir, current_position, move_history):
    if current_dir == 0:
        # to the north
        forhead_x = current_position[1]
        forhead_y = current_position[0] - 1
    elif current_dir == 1:
        # to the east
        forhead_x = current_position[1] + 1
        forhead_y = current_position[0]
    elif current_dir == 2:
        # to the south
        forhead_x = current_position[1]
        forhead_y = current_position[0] + 1
    elif current_dir == 3:
        # to the west
        forhead_x = current_position[1] - 1
        forhead_y = current_position[0]
    
    current_position = [forhead_y, forhead_x]
    move_history[forhead_y, forhead_x] = 2
    
    return current_position, move_history

def move_to_backward(current_dir, current_position, array_xy):
    if current_dir == 0:
        # to the south
        forhead_x = current_position[1]
        forhead_y = current_position[0] + 1
    elif current_dir == 1:
        # to the west
        forhead_x = current_position[1] - 1
        forhead_y = current_position[0]
    elif current_dir == 2:
        # to the north
        forhead_x = current_position[1]
        forhead_y = current_position[0] - 1
    elif current_dir == 3:
        # to the east
        forhead_x = current_position[1] + 1
        forhead_y = current_position[0]
    
    forhead_position = [forhead_y, forhead_x]
    
    forhead_land_sea = array_xy[forhead_y][forhead_x]
    
    back_move_ability = False
    
    if forhead_land_sea == 1:
        back_move_ability = False
        return current_position, back_move_ability
    
    elif forhead_land_sea == 0:
        back_move_ability = True
        return forhead_position, back_move_ability

move_cnt = 0

while True:
    for i in range(0, 4):
        current_dir = rotate_left(current_dir)
        move_ability = search_current_dir_move_ability(current_dir, current_position, array_xy, move_history)

        if move_ability:
            break
    
    if move_ability:
        print("current_dir is ", current_dir)
        current_position, move_history = move_to_currnet_dir(current_dir, current_position, move_history)
        move_cnt += 1
    else:
        print("current_dir is  ", current_dir)
        current_position, back_move_ability = move_to_backward(current_dir, current_position, array_xy)
        
        if back_move_ability == False:
            break
        else:
            move_cnt += 1
    
print(move_cnt)
```
